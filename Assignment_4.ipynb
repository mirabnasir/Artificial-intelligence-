```python

import pandas as pd
# Load the dataset
titanic_df = pd.read_csv("tested.csv")

# Display the first few rows of the dataset
titanic_df.head()
# Display the structure of the dataset
titanic_df.info()

# Statistical summary of the dataset
titanic_df.describe()

# Checking for null values
titanic_df.isnull().sum()
# Handling missing values
# For simplicity, let's fill missing 'Age' values with the median and drop rows with missing 'Embarked' and 'Cabin' values.
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)
titanic_df.drop(columns=['Cabin'], inplace=True)
titanic_df.dropna(subset=['Embarked'], inplace=True)

# Checking for duplicates
titanic_df.duplicated().sum()

# Dropping duplicates if any
titanic_df.drop_duplicates(inplace=True)

# Display the cleaned dataset structure
titanic_df.info()
import matplotlib.pyplot as plt
import seaborn as sns

# Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(titanic_df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
# Survival rate by gender
plt.figure(figsize=(10, 6))
sns.barplot(x='Sex', y='Survived', data=titanic_df, ci=None)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()
# Age vs. Fare
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=titanic_df)
plt.title('Age vs. Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived')
plt.show()
```
